```python
import re

file_path = r"C:\Users\netoo\OneDrive\Documentos\python projects\lei_8112\8112.txt"
substrings_to_remove = ["Redação", "Incluído", "VETADO", 'Vigência', 'Vide', 'Parágrafo','Revogado', 'Regulamento','Mantido', 'Revogada']

try:
    with open(file_path, 'r', encoding='utf-8') as file:
        file_contents = file.read()

    # Use regular expression to remove content inside parentheses for each substring
    for substring in substrings_to_remove:
        pattern = re.compile(fr'\({re.escape(substring)}[^)]*\)')
        file_contents = pattern.sub('', file_contents)

    # Write the modified content back to the file
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(file_contents)

    print(f"Content inside parentheses for '{', '.join(substrings_to_remove)}' has been removed.")
except UnicodeDecodeError:
    print("Unable to decode the file with the specified encoding.")
except Exception as e:
    print(f"An error occurred: {e}")
```
